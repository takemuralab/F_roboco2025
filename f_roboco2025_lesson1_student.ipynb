```python
# === 設定（必要に応じて編集）===
MODE = "generate"   # "generate" | "upload" | "drive"
CSV_NAME = "greenhouse_data.csv"
DRIVE_PATH = "/content/drive/MyDrive/datasets/greenhouse_data.csv"  # MODE="drive" の場合のパス

print("MODE =", MODE)
```

```python
import numpy as np
import pandas as pd

def generate_greenhouse_csv(path: str, n: int = 180, seed: int = 20251024):
    rng = np.random.default_rng(seed)
    dates = pd.date_range('2025-01-01', periods=n, freq='D')
    temperature = rng.normal(loc=24, scale=4, size=n).clip(8, 42)
    humidity = rng.normal(loc=68, scale=12, size=n).clip(25, 100)
    light = rng.normal(loc=16000, scale=4500, size=n).clip(2000, 35000)
    growth = 0.09*temperature + 0.00007*light - 0.018*np.abs(humidity-65) + rng.normal(0, 0.9, size=n)
    growth = np.clip(growth, 0, None)

    df = pd.DataFrame({
        "date": dates,
        "temperature": np.round(temperature, 2),
        "humidity": np.round(humidity, 1),
        "light": np.round(light, 0).astype(float),  # nullable int for NaN handling
        "growth": np.round(growth, 2),
    })
    # NaNs for practice
    df.loc[df.sample(frac=0.06, random_state=1).index, "humidity"] = np.nan
    df.loc[df.sample(frac=0.03, random_state=2).index, "light"] = np.nan

    df.to_csv(path, index=False)
    return df

if MODE == "generate":
    df_gen = generate_greenhouse_csv(CSV_NAME)
    print(f"Generated: {CSV_NAME}, shape={df_gen.shape}")
    display(df_gen.head())

```
